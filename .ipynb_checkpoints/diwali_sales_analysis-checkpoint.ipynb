**EXPLORATORY DATA ANALYSIS**

**Gender**

```python
#plotting a barchart for total count of genders
bx=sns.countplot(x='Gender' , data=df)

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_gen=df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
```

```python
sales_gen
```

```python
#plotting a bargraph for each  total count of genders vs their purchasing power

bx=sns.barplot(x='Gender' ,  y ='Amount' , data= sales_gen)
for bars in bx.containers:
    bx.bar_label(bars)
```

**from above graphs we can say that maximum buyer are female and also maximan puchasing amount of females are greater than man**

**Age**

```python
df.columns
```

```python
#plotting a barchart for total count of age group 
bx=sns.countplot(x='Age Group' , data=df , hue='Gender')

for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_gen=df.groupby(['Age Group' , 'Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
```

```python
sales_gen
```

```python
#plotting a bargraph for age group vs total amount


bx=sns.barplot(x='Age Group' ,  y ='Amount' , data= sales_gen  , hue='Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

**from above graph we can say that age group from 26-35 (females) are top buyer and also their purchasing power is greater
than other group**

**States**

```python
df.columns
```

```python
sales_state=df.groupby(['State'] , as_index=False)['Orders'].sum().sort_values(by='Orders' , ascending=False).head(10)
```

```python
sales_state
```

```python
#plotting a bargraph for top 10 states vs total orders

sns.set(rc={'figure.figsize':(15,5)})
bx=sns.barplot(x='State' ,  y ='Orders' , data= sales_state )
for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_state=df.groupby(['State'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False).head(10)
```

```python
#plotting a bargraph for top 10 states vs total amount

sns.set(rc={'figure.figsize':(15,5)})
bx=sns.barplot(x='State' ,  y ='Amount' , data= sales_state )
for bars in bx.containers:
    bx.bar_label(bars)
```

**From above graphs we can tell that Uttarpradesh have the most order as well as amount from other states following
Maharashtra and Karnataka . We can see that last 3 states for orders and amount changes in both the graphs**

**Marital Status**

```python
df.columns
```

```python
#plotting couontplot for marital status

sns.set(rc={'figure.figsize':(5,5)})
bx=sns.countplot(x='Marital_Status' , data=df)


for bars in bx.containers:
    bx.bar_label(bars)
```

```python
sales_gen=df.groupby(['Marital_Status' , 'Gender' ] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
#plotting a bargraph for age group vs total amount

sns.set(rc={'figure.figsize':(5,5)})
bx=sns.barplot(x='Marital_Status' ,  y ='Amount' , data= sales_gen  , hue='Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

**From above graphs we can see that most of the buyers are married (women) and they have high purchasing power**

**Occupation**

```python
df.columns
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.countplot(x='Occupation' , data=df)
```

```python
sales_gen=df.groupby(['Occupation' , 'Gender'] , as_index=False)['Amount'].sum().sort_values(by='Amount' , ascending=False)
#plotting a bargraph for occupation for each gender   vs total amount


bx=sns.barplot(x='Occupation' ,  y ='Amount' , data= sales_gen  , hue='Gender')
for bars in bx.containers:
    bx.bar_label(bars)
```

**From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector and in all of them females are the maximum buyer with maximum amount**

**Product Category**

```python
sns.set(rc={'figure.figsize':(25,5)})
ax = sns.countplot(data = df, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

**From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category**

```python
#top 10 prooducts sold by the company

sales_gen = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sales_gen
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = sales_gen, x = 'Product_ID',y= 'Orders')
```

**from above we can find top 10 sold product for the company**

**Conclusion**

*Married women age group 26-35 yrs from UP,  Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronics category*
